```python
import yfinance as yf

# Define the stock symbol and date range
sp500_tickers = [
    "AAPL", "MSFT", "AMZN", "GOOGL", "GOOG", "FB", "JPM", "TSLA", "V", "JNJ",
    "PG", "NVDA", "BRK.B", "HD", "MA", "UNH", "VZ", "PYPL", "DIS", "CRM",
    "INTC", "XOM", "CMCSA", "T", "WMT", "KO", "PFE", "MRK", "CSCO", "AAP",
    "COST", "NFLX", "CVX", "ABT", "MCD", "IBM", "GS", "SBUX", "CAT", "MO"
]

start_date = "2021-12-31" 
end_date = "2022-12-31"

ticker_dict = {}
quarterly_performance = {}

# Iterate through the list
for ticker in sp500_tickers:
    ticker_dict[ticker] = yf.download(ticker, start=start_date, end=end_date)
    
    try:
        temp_df = ticker_dict[ticker].copy()

        quarterly_data = temp_df['Adj Close'].resample('Q').ffill()

        quarterly_returns = quarterly_data.pct_change()

        # Store quarterly returns in the dictionary
        quarterly_performance[ticker] = quarterly_returns
        
    except Exception as e:
        print(f"Error processing data for {ticker}: {e}")

    # Store quarterly returns in the dictionary
    quarterly_performance[ticker] = quarterly_returns
```

```python
#Adding the Daily ROI column
for df_idx in sp500_tickers:
    target_df = ticker_dict[df_idx]
    target_df['Daily_ROI'] = (target_df['Adj Close'] / target_df['Adj Close'].shift(1) - 1) * 100

#Testing to make sure the Daily ROI is correct
#Keep in mind the first value is none because there is no value beforehand
ticker_dict['AAPL'].head()
```

```python
#using pickle to export
import pickle

with open('ticker_prices.pkl', 'wb') as pickle_file:
    pickle.dump(ticker_dict, pickle_file)
    
with open('Quarter_prices.pkl', 'wb') as pickle_file:
    pickle.dump(quarterly_performance, pickle_file)
```
